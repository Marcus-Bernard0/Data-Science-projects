```python
import pandas as pd
```

```python
df = pd.read_csv('dados/imoveis.csv', sep = ';')
df.head()
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.rename(columns={
    'Localização': 'localizacao',
    'Endereço': 'endereco',
    'Descrição': 'descricao',
    'Tamanho': 'area_m2',
    'Quartos': 'quartos',
    'Banheiros': 'banheiros',
    'Vagas Garagem': 'vagas_garagem',
    'Valores': 'valor',
    'Links': 'link'
}, inplace=True)
```

```python
df.info()
```

## Transformando colunas

```python
df['area_m2'].unique()
```

```python
df.quartos = df.quartos.astype('int64')
```

```python
df.quartos.unique()
```

```python
df.quartos.info()
```

```python
df.columns
```

```python
df.banheiros.unique()
```

```python
df.banheiros = df.banheiros.astype('int64')
```

```python
df.vagas_garagem.unique()
```

```python
df.vagas_garagem = df.vagas_garagem.astype('int64')
```

## Lendo os novos dados depois de tratado no Power Query

```python
df_oficial = pd.read_csv('dados/imoveis_tratados_oficial.csv', sep=';')
```

```python
df_oficial.head()
```

```python
df_oficial[['bairro', 'cidade']] = df_oficial.localizacao.str.split(',', 1, expand=True)
```

```python
df_oficial.head()
```

```python
df_oficial.loc[df_oficial['cidade'].isnull()]
```

```python
df_oficial.cidade = df_oficial.cidade.fillna(df_oficial['bairro'])
```

```python
df_oficial.isna().sum()
```

```python
df_oficial.drop(['localizacao'], axis=1, inplace=True)
```
