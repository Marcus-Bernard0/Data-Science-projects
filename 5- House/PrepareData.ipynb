```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = pd.read_csv('train.csv')
    return data
dfTrain = load_data()
```

```python
def infoData():
    desc = dfTrain.describe()
    inf = dfTrain.info()
    return display(desc, inf)
infoData()
```

```python
dfTrain.hist(bins=50, figsize=(20,15), color='blue', lw=1.5, alpha=0.5, grid=False, edgecolor='black', linewidth=1.5, rwidth=1.9)
plt.show()
```
