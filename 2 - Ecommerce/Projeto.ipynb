# Análise Exploratória dos dados

### Qual categoria de produto tem o preço maior?

```python
dfProductsMerge = pd.merge(dfProducts, dfOdersItens, how='inner', on='product_id')
dfProductsMerge.head()
```

```python
dfProductPrice = dfProductsMerge.groupby('product_category_name', as_index=False)['price'].mean()
dfProductPrice = pd.DataFrame(dfProductPrice)
dfProductPrice.head()
```

```python
dfProductPrice = dfProductPrice.sort_values('price', ascending=False)
dfProductPrice
```

#### 10 produtos mais caros (em média)

```python
df10Products = dfProductPrice.iloc[0:10]
df10Products
```

```python
df10Products
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_palette('Blues_r')
sns.set_style('ticks')
```

```python
def graphproducts ():
  plt.figure(figsize = (12, 8))
  ax = sns.barplot(x = 'product_category_name', y = 'price', data = df10Products, 
                  order = df10Products.sort_values('price', ascending = False).product_category_name, 
                  palette='Blues_r')
  plt.xticks(rotation = 90)
  ax.set_title("Análise das 10 categorias de produtos mais caros", fontsize = 18,  x = 0.15, y = 1.05)
  ax.set_ylabel("Preços")
  ax.set_xlabel("Categoria", fontsize = 12 )
  for i in ax.containers:
        ax.bar_label(i,)
  plt.show()
```

```python
graphproducts()
```

```python
def pcs():
    ax = sns.boxplot(x = 'price', y = 'product_category_name', 
                data = dfProductsMerge.query('product_category_name == "pcs"'),
                palette = 'Blues_r')
    ax.set_title("Boxplot da categoria Pcs", fontsize = 18, x = 0.15, y = 1.05)
    ax.figure.set_size_inches(12, 6)
    plt.show()
```

```python
def utd ():
    ax = sns.boxplot(x = 'price', y = 'product_category_name', orient='h',
                data = dfProductsMerge.query('product_category_name == "utilidades_domesticas"'),
                palette = 'Blues_r',  width=0.4)
    ax.figure.set_size_inches(12, 6)
    ax.set_title("Boxplot da categoria Utilidades domésticas", fontsize = 18, x = 0.15, y = 1.05)
    plt.show()
```

#### Estatísticas descritivas Utilidades domésticas e PCs


```python
dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price'].min()
```

```python
valor = dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price']
def estatistica_utd() : 
    count, mean, std, min,Q1, Q2, Q3,max = valor.describe()
    amplitude = dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price'].max() - dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price'].min()
    cv = (std/mean) * 100
   
    print('Estatísticas categoria útilidades domésticas:')
    print(f'Q1: {Q1}')
    print(f'Q2 : {Q2}')
    print(f'Q3 : {Q3}')
    print(f'Valor máximo {max}')
    print(f'Valor mínimo {min}')
    print(f'A média para essa categoria é R$ {round(mean, 2)}' )
    print(f'Com desvio padrão de {round(std, 2)}')
    print(f'Com amplitude de {round(amplitude, 2)}')      
    print(f'Com coeficiente de variação {round(cv, 2)}')
    
```

```python
estatistica_utd()
```

Estatísticas descritivas Agro indústria.

```python
def estatistica_pcs() : 
    valor = dfProductsMerge.query('product_category_name == "pcs"')['price']
    count, mean, std, min,Q1, Q2, Q3,max = valor.describe()
    amplitude = dfProductsMerge.query('product_category_name == "pcs"')['price'].max() - dfProductsMerge.query('product_category_name == "pcs"')['price'].min()
    cv = (std/mean) * 100
    print('Estatísticas categoria PCS:')
    print(f'Q1: {Q1}')
    print(f'Q2 : {Q2}')
    print(f'Q3 : {Q3}')
    print(f'Valor máximo {max}')
    print(f'Valor mínimo {min}')
    print(f'A média para essa categoria é R$ {round(mean, 2)}' )
    print(f'Com desvio padrão de {round(std, 2)}')
    print(f'Com amplitude de {round(amplitude, 2)}')      
    print(f'Com coeficiente de variação {round(cv, 2)}')
```

```python
estatistica_pcs()
```
