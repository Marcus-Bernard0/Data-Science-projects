### Qual categoria de produto tem o preço maior?

Estatística descritiva sobre essa análise



```python
from operator import index


dfProductsMerge = pd.merge(dfProducts, dfOdersItens, how='inner', on='product_id')
dfProductsMerge.head()
```

```python
dfProductPrice = dfProductsMerge.groupby('product_category_name', as_index=False)['price'].mean()
dfProductPrice = pd.DataFrame(dfProductPrice)
dfProductPrice.head()
```

```python
dfProductPrice.groupby(['product_category_name']).price.agg(['max', 'min', 'mean', 'median'])
```

```python
import matplotlib.pyplot as plt
```

```python

sns.set_palette('Blues_r')
sns.set_style('ticks')
```

```python
ax = sns.barplot(x='product_category_name', y='price', data=dfProductPrice.loc[0:10], palette='Blues_r')
plt.xticks(rotation=90)
ax.set_title("Preços médios dos 10 produtos mais caros")
ax.set_xlabel("Categoria")
ax.set_ylabel("Preço")
plt.show()
```
