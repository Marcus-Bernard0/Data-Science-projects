### Qual categoria de produto tem o preço maior?

Estatística descritiva sobre essa análise



```python
dfProductsMerge = pd.merge(dfProducts, dfOdersItens, how='inner', on='product_id')
dfProductsMerge.head()
```

```python
dfProductPrice = dfProductsMerge.groupby('product_category_name', as_index=False)['price'].mean()
dfProductPrice = pd.DataFrame(dfProductPrice)
dfProductPrice.head()
```

```python
dfProductPrice.groupby(['product_category_name']).price.agg(['max', 'min', 'mean', 'median'])
```

```python
import matplotlib.pyplot as plt
```

```python

sns.set_palette('Blues_r')
sns.set_style('ticks')
```

```python
df10Products = dfProductPrice.loc[0:10]
```

```python
def graphproducts ():
  plt.figure(figsize = (12, 8))
  ax = sns.barplot(x = 'product_category_name', y = 'price', data = df10Products, 
                  order = df10Products.sort_values('price', ascending = False).product_category_name, 
                  palette='Blues_r')
  plt.xticks(rotation = 90)
  ax.set_title("Análise das 10 categorias de produtos mais caros", fontsize = 18)
  ax.set_ylabel("Preços")
  ax.set_xlabel("Categoria", fontsize = 12 )
  plt.show()
```

```python
graphproducts()
```

```python
dfProductPrice.describe()
```

```python
dfMeanCat = pd.DataFrame(dfProductPrice.groupby('product_category_name', as_index = False)['price'].mean().round(2))
dfMeanCat
```

**Insight 1 : A categoria agro indústria tem a maior média dos preços. Ou seja, produtos mais caros. (Em média)**
