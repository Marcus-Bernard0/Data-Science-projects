## Introdução

## Modelo de Machine Learning

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from numpy import random
```

```python

dfFeatures = df.drop(columns= ['id', 'diagnosis'])
dfTarget = df['diagnosis']

treino_x, teste_x, treino_y, teste_y = train_test_split(dfFeatures, dfTarget, test_size=0.3, random_state=42)
classificador = RandomForestClassifier(n_estimators=100)
classificador.fit(treino_x, treino_y)

selecionador_rfecv = RFECV(estimator = classificador, cv = 10, scoring = "accuracy", step = 1)
selecionador_rfecv.fit(treino_x, treino_y)
treino_rfecv = selecionador_rfecv.transform(treino_x)
teste_rfecv = selecionador_rfecv.transform(teste_x)
classificador.fit(treino_rfecv, treino_y)

print("Resultado da classificação %.2f%%" % (classificador.score(teste_rfecv, teste_y)* 100))


```

```python
from sklearn.metrics import classification_report

report = classification_report(teste_y, classificador.predict(teste_rfecv))
print(report)
```

As colunas selecionadas pelo modelo

```python
treino_x.columns[selecionador_rfecv.support_]
```

## Classificador bobo

Esse é um classificador bobo que avalia nossa base line. Ou seja, o modelo bobo alcançou 63.16% de acurácia. Entretanto, o modelo desenvolvido alcançou 96.47% de acurácia.
O modelo construído é extremamente superior ao modelo bobo.

```python
from sklearn.dummy import DummyClassifier

SEED = 42
random.seed(SEED)

classificador_dummy = DummyClassifier(strategy="most_frequent")
classificador_dummy.fit(treino_x, treino_y)
classificador_dummy.score(teste_x, teste_y)
print(f'Acurácia dummy: {round(classificador_dummy.score(teste_x, teste_y)*100, 2)}%')
```
